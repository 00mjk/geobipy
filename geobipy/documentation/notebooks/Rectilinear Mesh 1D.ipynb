```python
%matplotlib notebook
%load_ext autoreload
%autoreload 2
from geobipy import StatArray
from geobipy import RectilinearMesh1D
import matplotlib.pyplot as plt
%matplotlib notebook
import numpy as np
```

```python
x = StatArray(np.cumsum(np.arange(0.0, 10.0)), 'Depth', 'm')
```

```python
rm = RectilinearMesh1D(cellEdges=x)
```

```python
rm.cellCentres
```

```python
rm.cellEdges
```

```python
rm.internalCellEdges
```

```python
rm.cellWidths
```

```python
plt.figure()
rm.plotGrid()
```

```python
plt.figure()
arr = StatArray(np.random.randn(rm.nCells), "Name", "Units")
rm.pcolor(arr, grid=True)
```

### Slicing into a 1D Rectilinear Mesh

```python
rm2 = rm[:-1]
arr = StatArray(np.random.randn(rm2.nCells), "Name", "Units")
plt.figure()
rm2.pcolor(values = arr, grid=True)
```
