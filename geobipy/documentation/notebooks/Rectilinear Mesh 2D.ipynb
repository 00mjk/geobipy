```python
%matplotlib notebook
%load_ext autoreload
%autoreload 2
from geobipy import StatArray
from geobipy import RectilinearMesh2D
import matplotlib.pyplot as plt
%matplotlib notebook
import numpy as np
```

## 2D mesh with regular bins

```python
x = StatArray(np.arange(10.0), 'Easting', 'm')
y = StatArray(np.arange(10.0), 'Northing', 'm')
```

```python
rm = RectilinearMesh2D(xCentres=x, yCentres=y)
```

```python
a = np.linspace(-1.0, 11.0, 100)
b = np.linspace(-1.0, 11.0, 100)
rm.cellIndices(a, b)
```

```python
rm.cellIndices(a, b, clip=True)
```

```python
rm.cellIndices(a, b, trim=True)
```

```python
plt.figure()
rm.plotGrid(flipY=False)
```

#### Slice into the mesh

```python
plt.figure()
rm[:9, :].plotGrid()
```

## 2.5D mesh with regular spacing in x-y and irregular in z

```python
z = StatArray(np.cumsum(np.arange(15.0)), 'Depth', 'm')
rm = RectilinearMesh2D(xCentres=x, yCentres=y, zCentres=z)
```

```python
plt.figure()
rm.plotGrid(xAxis='r', flipY=True)
```

```python
a = np.repeat(np.arange(1.0, np.float(rm.x.nCells+2))[:, np.newaxis], rm.z.nCells, 1).T
```

```python
a
```

### Compute the mean over an interval for the mesh.

```python
rm.intervalStatistic(a, [6.8, 12.4], 0, statistic='mean')
```

### Compute the mean over multiple intervals for the mesh.

```python
rm.intervalStatistic(a, [6.8, 12.4, 20.0, 40.0], 0, statistic='mean')
```

```python
rm.intervalStatistic(a, [2.8, 4.2], 1, statistic='mean')
```

```python
rm.intervalStatistic(a, [2.8, 4.2, 5.1, 8.4], 1, statistic='mean')
```

```python
plt.figure()
rm.pcolor(arr, xAxis='r', grid=True, flipY=True)
```
